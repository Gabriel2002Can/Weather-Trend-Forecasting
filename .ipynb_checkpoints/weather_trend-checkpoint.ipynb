## Feature Engineering

```python
filtered_df = filtered_df.drop(columns=["location_name", "country", "timezone", 'wind_direction'])
```

These columns don't add relevant information. location_name, country and timezone are all conveying geographical information, something that latitude and longitude already do, and wind direction is already present on the wind degree column.

```python
time_features = ["sunrise", "sunset", "moonrise", "moonset"]
```

```python
def cyclic_encode(series, period):
    radians = 2 * np.pi * series / period

    sin = np.sin(radians)
    cos = np.cos(radians)

    return sin, cos
```

```python
def transform_data_set(dataframe, property_name, period):
    
    dataframe[f"{property_name}_sin"], dataframe[f"{property_name}_cos"] = cyclic_encode(
    dataframe[property_name],
    period
    )
    
    dataframe = dataframe.drop(columns=[property_name])
    
    return dataframe
```

These functions are transforming a numeric column that represents cyclic data, mostly time related, into two new columns called column_sin and column_cos. The objective is to use sine and cosine to convey time in a 2D coordinate system, representing a cyclic relationship, something specially normal in time features.

### Parsing Datetime

```python
filtered_df["last_updated"] = pd.to_datetime(filtered_df["last_updated"])
```

```python
filtered_df["year"] = filtered_df["last_updated"].dt.year

filtered_df["year"].value_counts().index
```

```python
filtered_df["month"] = filtered_df["last_updated"].dt.month
```

```python
filtered_df["day_of_year"] = filtered_df["last_updated"].dt.dayofyear
```

```python
filtered_df["hour"] = filtered_df["last_updated"].dt.hour
```

```python
filtered_df = filtered_df.drop(columns=["last_updated"])
```

### Transforming Cyclic Features

```python
for time in time_features:
    parsed_time = pd.to_datetime(
        filtered_df[time],
        format="%I:%M %p",
        errors="coerce"
    )

    filtered_df[time] = (
        parsed_time.dt.hour +
        parsed_time.dt.minute / 60
    )

    filtered_df = transform_data_set(filtered_df, time, 24)
```

```python
moon_cols = [
    "moonrise_sin", "moonrise_cos",
    "moonset_sin", "moonset_cos"
]

filtered_df[moon_cols] = filtered_df[moon_cols].fillna(0)
```

Adjusting moonrise values to NOT include NaN values, otherwise the Logistic Regression model can fail.

```python
filtered_df = transform_data_set(filtered_df, "wind_degree", 24)
```

```python
filtered_df = transform_data_set(filtered_df, "hour", 24)
```

```python
filtered_df = transform_data_set(filtered_df, "month", 12)
```

```python
filtered_df = transform_data_set(filtered_df, "day_of_year", 365)
```

## Splitting the Data

```python
X = filtered_df.drop('y', axis=1)

y = filtered_df['y']
```

```python
# Splitting the train and test subsets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42 )
```

Splitting the data, using a 25% proportion for the test size and using the 42 seed as random state.

## Normalizing

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

This is specially necessary for the Logistic Regression model.

## Defining and Training Models

### Defining Models

```python
logistic_regressor = LogisticRegression()

random_forest_model = RandomForestClassifier()

xgboost_model = XGBClassifier()

ensemble = VotingClassifier(
    estimators=[
        ('lr', logistic_regressor),
        ('rf', random_forest_model),
        ('xgb', xgboost_model)
    ],
    voting='soft',
    weights=[1, 3.5, 5]
)
```

Here, three models are being defined:

- Logistic Regressor
- Random Forest
- XGBoost

Considering that the models score diferrently, a weight was implemented giving the better models more weight in the voting process. 

### Training Models

```python
models = {
    'lr': logistic_regressor,
    "rf": random_forest_model,
    "xgb": xgboost_model
}

for name, model in models.items():

    model.fit(X_train_scaled, y_train)

    labels = sorted(np.unique(y_train))

    # TRAIN
    y_train_result = model.predict(X_train_scaled)
    
    accuracy_training = accuracy_score(y_train, y_train_result) * 100
    classification_training = classification_report(y_train, y_train_result)
    conf_matrix_training = confusion_matrix(y_train, y_train_result)

    print(f'Accuracy Training Set {name} model: {accuracy_training}%\n')
    print(f'Classification Report Training Set {name} model: \n{classification_training}')

    cm = confusion_matrix(y_train, y_train_result, normalize="true", labels=labels)
    plt.figure(figsize=(22, 18))
    sns.heatmap(
        cm,
        cmap="Blues",
        xticklabels=labels,
        yticklabels=labels,
        annot=False
    )
    plt.title("Training Confusion Matrix")
    plt.show()

    # TEST
    y_test_result = model.predict(X_test_scaled)
    
    accuracy_test = accuracy_score(y_test, y_test_result) * 100
    classification_test = classification_report(y_test, y_test_result)
    conf_matrix_test = confusion_matrix(y_test, y_test_result)

    print(f'Accuracy Test Set {name} model: {accuracy_test}%\n')
    print(f'Classification Report Test Set {name} model: \n{classification_test}')

    cm = confusion_matrix(y_test, y_test_result, normalize="true", labels=labels)
    plt.figure(figsize=(22, 18))
    sns.heatmap(
        cm,
        cmap="Blues",
        xticklabels=labels,
        yticklabels=labels,
        annot=False 
    )
    plt.title("Testing Confusion Matrix")
    plt.show()
```

## Ensemble

```python
cross_val_score(ensemble, X_train,y_train, cv=3).mean()
```

This is the accuracy for the cross validation accuracy, considering the ensemble model.

```python
ensemble.fit(X_train, y_train)
```
