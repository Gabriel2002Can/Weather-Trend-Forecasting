# Weather Trend Forecasting Assignment

## Importing Modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import seaborn as sns
import squarify
```

## Importing Data

```python
df = pd.read_csv("GlobalWeatherRepository.csv")

df
```

## Data Analysis

```python
df.columns
```

Remove inrelevent features: **temperature_fahrenheit, feels_like_fahrenheit, visibility_miles, gust_mph**

```python
df = df.drop(columns=["temperature_fahrenheit", "feels_like_fahrenheit", "visibility_miles", "gust_mph", "precip_in", "pressure_in"])

len(df.columns)
```

```python
df
```

```python
df.describe()
```

```python
df = df.rename(columns = {
    'condition_text':'y'
})
```

```python
df[df.isna().any(axis=1)]
```

```python
df["y"] = df["y"].str.strip().str.lower()
```

```python
top_y = df["y"].value_counts().nlargest(15).index

filtered_df = df.copy()

# If NOT on the top y, is rest
filtered_df["y"] = filtered_df["y"].apply(
    lambda x: x if x in top_y else "rest"
)

df["y"].value_counts().nlargest(15)
```

```python
df["y"]
```

```python
weather_count = filtered_df["y"].value_counts()

fig, ax = plt.subplots(1, 1, figsize=(25,10))

sns.barplot(y=weather_count.values, x=weather_count.index, hue=weather_count.index, ax=ax, palette="viridis")
ax.set_title("Distribution of Weathers")
ax.set_xlabel("Weather")
ax.tick_params(axis='x', labelrotation=90)
ax.set_ylabel("Number of records")
```

```python
countries_grouped = df.groupby("country").size()

countries_grouped.plot(kind="barh", figsize=(25,50))
```

```python
fig, ax = plt.subplots(1,1, figsize=(20,20))

sns.violinplot(data=filtered_df, y="temperature_celsius", hue="y" , ax=ax, palette="viridis")
ax.set_title("Distribution of Temperature (°C)")
ax.set_xlabel("Weather")
ax.tick_params(axis='x', labelrotation=90)
ax.set_ylabel("Temperature (°C)")
```

```python
fig, ax = plt.subplots(1,1, figsize=(20,20))

sns.boxplot(data=filtered_df, y="y", x="precip_mm", ax=ax, palette="viridis")

ax.set_title("Distribution of Precipitation (mm)")
ax.set_ylabel("Weather")
ax.set_xscale("log")
ax.set_xlabel("Precipitation (mm)")
```
